## 2. Pass/Fail Analyzer

```python
student_marks = [45, 78, 90, 33, 60]  # Marks of indivisual students are in the list
Pass = 0
Fail = 0
for marks in student_marks:
    if marks >= 50:
        Pass += 1
    else:
        Fail += 1

print("Total number of pass students: ",Pass)
print("Total number of fail students: ",Fail)
```

## 3. Simple Data cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]

def clean_data(lists):
    cleaned = []
    for i in lists:
        a = (i.replace(" ","")).lower()
        cleaned.append(a)
    return cleaned
    
print("Cleaned List: ",clean_data(names))
```

## 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for i in messages:
    if len(i) <= 10:
        print(f"Length of Message : '{i}' : {len(i)}")
    else:
        print(f"Length of Long Messages : '{i}' : {len(i)} [Flagged]")
            
```

## 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

def error_detector(logs):
    errors = 0
    for msg in logs:
        if msg == "ERROR":
            errors += 1
    return errors

print("Total Error Count: ", error_detector(logs))
```
